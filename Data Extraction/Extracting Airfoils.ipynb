```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os

```

```python
URL= "http://www.airfoiltools.com/search/airfoils"
```

```python
HEADER= ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'})
```

```python
webpage= requests.get(URL, headers=HEADER)
```

```python
webpage
```

```python
soup= BeautifulSoup(webpage.content, "html.parser")
```

```python
td_links= soup.find_all("td", attrs={'class':"link"})
```

```python
td_links
```

```python
airfoil=[]
for td in td_links:
    a_tag = td.find("a")
    if a_tag:
        link = a_tag.get("href")
        airfoil.append(link)        
print("Link:", airfoil)

```

```python
print(airfoil)
```

```python
base_url="http://www.airfoiltools.com"
airfoil_links= [base_url + link for link in airfoil]
print("Airfoil Link", airfoil_links)

```

```python
replace_word= "details"
replacement_word= "seligdatfile"
list_modified_links= [link.replace(replace_word, replacement_word) for link in airfoil_links]
list_modified_links

folder = r"C:\Users\shrey\OneDrive\Desktop\Phinal year project\scrapping\airfoil dat files\new"

for index, modified_link in enumerate(list_modified_links):
    filename= modified_link.split('=')[-1] + '.dat'
    print(filename)
    filepath= os.path.join(folder, f"{filename}")
#     print(filepath)
    
    if os.path.exists(filepath):
        print(f"File already exists: {filepath}. Skipping...")
        continue

    print(f"Downloading and saving modified_link to {filepath}...")
    
    response = requests.get(modified_link)
    content = response.text
#     print(content)
    with open(filepath, 'w') as file:
        file.write(content)
    
    print(f"modified_link saved to {filepath}")
```
